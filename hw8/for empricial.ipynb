```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load returns for raw and tile features
raw_returns = np.loadtxt("/Users/leesunny/PycharmProjects/10601 ML/returns_raw_1.txt")  # Replace with your actual file path
tile_returns = np.loadtxt("/Users/leesunny/PycharmProjects/10601 ML/returns_tile_1.txt")  # Replace with your actual file path

# Define function to calculate rolling mean
def rolling_mean(data, window):
    return pd.Series(data).rolling(window=window).mean()

# Plot for Raw Features
plt.figure(figsize=(10, 6))
plt.plot(raw_returns, label="Raw Returns", alpha=0.7)
plt.plot(rolling_mean(raw_returns, 25), label="Rolling Mean (25 episodes)", linewidth=2)
plt.title("Q-Learning with Raw Features")
plt.xlabel("Episode")
plt.ylabel("Total Return")
plt.legend()
plt.grid(True)
plt.savefig("raw_features_plot.png")  # Save the plot
plt.show()

# Plot for Tile Features
plt.figure(figsize=(10, 6))
plt.plot(tile_returns, label="Tile Returns", alpha=0.7)
plt.plot(rolling_mean(tile_returns, 25), label="Rolling Mean (25 episodes)", linewidth=2)
plt.title("Q-Learning with Tile Features")
plt.xlabel("Episode")
plt.ylabel("Total Return")
plt.legend()
plt.grid(True)
plt.savefig("tile_features_plot.png")  # Save the plot
plt.show()
```
